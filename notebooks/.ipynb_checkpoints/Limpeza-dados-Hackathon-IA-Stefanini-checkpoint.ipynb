### Colunas 

Com excessão de `Date` todas as outras colunas parecem estar com o tipo correto. 

`Date` está como object, uma string, se quisermos utilizar essa informação adequadamente, teremos que transformar em Datetime 


No final da data, há uma string "A.M" que indica um período. Vamos retirá-la antes de fazer a transformação

### Documentação

[pd.to_datetime](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html)

```python
def remove_string(row):
    return row.strip('A.M.')
```

```python
trade['Date'] = trade['Date'].apply(remove_string)
trade['Date'] = trade['Date'].apply(pd.to_datetime)
```

```python
trade.dtypes
```

Agora podemos acessar atributos de tempo da coluna `Date`. Vamos partir para a visualização, ela pode ajudar a entender mais os dados e a identificar outliers e anomalia na informaçao



**Importante**

Para fazer uma visualização dos dados por mês, terei que adicionar essa coluna no dataframe.


**A adição da coluna Month já é um passo de Engenharia de Características**

```python
trade['Month'] = trade['Date'].apply(lambda x: x.month)
trade.head()
```

Feita a Adição, agora posso me perguntar o seguinte: 

> Como é o comportamento da moeda em relação aos Meses?

```python
mean_of_trades = trade.groupby('Month').mean().reset_index()
mean_of_trades.head()
```

## Esse gráfico mostra o comportamento da moeda durante toda sua existência. Pois na média definida acima, é levado em consideração todos os anos


Pela razão da `manutenibilidade`, criei uma função pra plotar os gráficos que utilizarei com frequência

```python
def plot_subplot_bonitao(x, y1, y2, data, hue=None, rotation=30):
    fig, axes = plt.subplots(1, 2, figsize=(9,5), sharex=True)
    _ = sns.barplot(x, y1, data=data, hue=hue, ax=axes[0], palette='muted')
    _ = _.set_xticklabels(_.get_xticklabels(), rotation=rotation)

    _ = sns.barplot(x, y2, data=data, hue=hue, ax=axes[1], palette='Blues_d')
    _ = _.set_xticklabels(_.get_xticklabels(), rotation=rotation)
```

```python
plot_subplot_bonitao(x='Month', y1="Open", y2='Close', data=mean_of_trades)
```

### Vamos tentar achar uma variação ao longo dos anos


Para isso, precisarei de uma coluna com os anos. 

```python
trade['Year'] = trade['Date'].apply(lambda x: x.year)
trade.head()
```

### Reset_index()

Uso o [reset_index()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.reset_index.html) para fazer com que o dataframe ""drope"" os index de multinível depois que eu chamo o groupby. 

Observe como seria sem, e com, o reset_index():

```python
trade.groupby('Year').mean().head()
```

```python
mean_of_trades_years = trade.groupby('Year').mean().reset_index()
mean_of_trades_years.head()
```

```python
plot_subplot_bonitao(x='Year', y1="Open", y2='Close', data=mean_of_trades_years)
```

A média do "Open" e do "Close" durante o ano é muito similar. Quase como dizer que:

> O preço que abriu vai ser mais ou menos o preço que vai fechar

```python
plot_subplot_bonitao(x='Year', y1="High", y2='Low', data=mean_of_trades_years)
```

### Vamos ver se o período dentro do mês influenciava algo. 


Para isso, precisaremos de uma coluna com os **dias** do mês

```python
trade['Day'] = trade['Date'].apply(lambda x: x.day)
trade.head(20)
```

## Há muitos dias!

bem, há 30, ou 31, dias em um mês. Talvez saber o quanto variou para exatamente cada dia não seja uma estratégia interessante. Pode ser muito especialista. Pra ficar mais interessante, vamos transformar os dias em um período. Ou seja, realizar um [Binning](https://en.wikipedia.org/wiki/Data_binning). 

Nesse binning, estou considerando que até o dia 16 é o início do mês e todos os dias após, é o final



Podemos fazer isso com o uso do [pd.cut](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.cut.html)

```python
trade['periodo do mês'] = pd.cut(trade['Day'], 2, labels=['inicio', 'fim'])
trade.head(20)
```

```python
plot_subplot_bonitao(x='Year', y1="Open", y2='Close', data=mean_of_trades_years)
```

### Vamos ver a variação da moeda no início e final do mês ao longo dos anos

```python
year_month_period = trade.groupby(['Year','periodo do mês']).mean().reset_index()
year_month_period.head()
```

```python
plot_subplot_bonitao(x='Year', y1='Open', y2='Close', hue='periodo do mês', data=year_month_period)
```

### Vamos ver a variação da moeda no início e final do mês visto pelos meses ao longo dos anos

```python
month_month_period = trade.groupby(['Month','periodo do mês']).mean().reset_index()
month_month_period.head()
```

```python
plot_subplot_bonitao(x='Month', y1='Open', y2='Close', hue='periodo do mês', data=month_month_period)
```

## Justificativa

Para realizar uma regressão com as variações das cotações da moeda DinoCoin, onde possamos analisar as variações da cotação da moeda. Vamos utilizar somente as informações de dada. Isso se eu quiser saber a variação dela dado determinado período. 

Essa decisão parte por imaginar que não se sabe os valores de abertura e fechamento da moeda para uma data futura.


Uma vez que desejamos saber qual foi a variação da moeda, devemos calcular a variação da moeda para ser uma `feature` do nosso modelo

```python
trade_tratado['Variacao'] = trade_tratado['Close'] - trade_tratado['Open']
trade_tratado.head()
```

```python
columns_to_stay = ['Month', 'Year', 'periodo do mês', 'Variacao']
trade_tratado = trade_tratado[columns_to_stay]
```

Massa! Mas `periodo do mês` é uma coluna que indica categorias vamos ter que transformar isso em números, pois os modelos do Sklearn não entendem string. Podemos fazer isso de diversas maneiras, entre elas:

1) Transformar pelo pandas. Criamos uma função que transforma "Sim" e "Não" em 0 e 1, respectivamentne

2) Utilizar o [LabelEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html) do Scikit lá na hora de fazer o modelo

3) Usar o [pd.get_dummies](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html) e transformar em duas novas colunas através da técnica do [OneHotEncoding](https://hackernoon.com/what-is-one-hot-encoding-why-and-when-do-you-have-to-use-it-e3c6186d008f)

4) Fazer o OneHotEncoding com o Sklearn usando o [OneHotEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html#sklearn.preprocessing.OneHotEncoder)
